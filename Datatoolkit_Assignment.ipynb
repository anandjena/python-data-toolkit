1.What is NumPy, and why is it widely used in Python?

2. How does broadcasting work in NumPy?

3.What is a Pandas DataFrame?

4. Explain the use of the groupby() method in Pandas.
- The groupby() method in Pandas is used to split a DataFrame into groups based on one or more column values, allowing you to perform aggregations or transformations on each group independently. It follows a split-apply-combine strategy: it splits the data into groups, applies a function (like mean(), sum(), count(), etc.) to each group, and then combines the results into a new DataFrame or Series. This is particularly useful for summarizing and analyzing data by categories or classes, such as computing the average salary by department or the total sales by region. The groupby() method is powerful and flexible, enabling operations like grouping by multiple columns, applying custom functions, filtering groups, or even transforming each group’s data without aggregation. Overall, it is a key tool in data analysis that simplifies complex data summarization tasks.

5.Why is Seaborn preferred for statistical visualizations?

6.What are the differences between NumPy arrays and Python lists?
- NumPy arrays and Python lists both store collections of data, but they differ significantly in functionality, performance, and usage. NumPy arrays are homogeneous, meaning all elements must be of the same data type, which allows for optimized memory usage and faster computation. In contrast, Python lists are heterogeneous and can hold different data types, but this flexibility comes at the cost of speed and memory efficiency. NumPy arrays support vectorized operations, allowing element-wise arithmetic and mathematical functions without writing explicit loops, whereas Python lists require manual iteration for such tasks. Additionally, NumPy arrays are ideal for multidimensional data and offer powerful capabilities for linear algebra, statistics, and broadcasting, which are not natively supported by Python lists. Overall, while Python lists are more flexible for general-purpose use, NumPy arrays are preferred for numerical and scientific computing due to their speed, efficiency, and rich functionality.

7. What is a heatmap, and when should it be used?
- A heatmap is a data visualization technique that uses color to represent values in a matrix or 2D dataset. Each cell in the heatmap corresponds to a value, and the color intensity indicates the magnitude of that value — typically, darker or more vivid colors represent higher values, and lighter or muted colors represent lower ones.

8.What does the term “vectorized operation” mean in NumPy?
- In NumPy, a vectorized operation refers to performing element-wise operations on entire arrays without the need for explicit loops in Python. This means that you can apply arithmetic or mathematical operations directly to NumPy arrays, and the operation is automatically applied to each element efficiently under the hood using optimized C code.

9.How does Matplotlib differ from Plotly?
- Matplotlib and Plotly are both popular Python libraries for data visualization, but they differ significantly in functionality and use cases. Matplotlib is a static, low-level plotting library that offers fine-grained control over every aspect of a plot, making it highly customizable and suitable for traditional, publication-quality graphs. However, its visualizations are typically static (non-interactive) and may require more code for complex plots. In contrast, Plotly is designed for interactive, web-based visualizations and supports features like zooming, hovering, and dynamic updates out of the box. Plotly is often easier to use for creating dashboards and interactive data exploration tools, and it integrates well with web technologies and frameworks like Dash. While Matplotlib is ideal for scientific and academic work requiring precision and control, Plotly is better suited for modern data storytelling, especially when interactivity is needed.

12.What is the purpose of the describe() function in Pandas?

It returns key descriptive statistics such as count, mean, standard deviation, minimum, maximum, and the quartiles (25%, 50%, 75%), giving a concise overview of the data’s distribution and spread. This function helps you understand the central tendency, variability, and range of your data at a glance, which is essential for initial exploratory data analysis and detecting anomalies or data quality issues. 15.How does NumPy handle multidimensional arrays?

Each dimension is called an axis, and the shape of the array is described by a tuple indicating the size along each axis. 17.Explain the difference between apply() and map() in Pandas.
- In Pandas, map() and apply() both allow you to apply functions to data, but they differ mainly in their scope and flexibility. The map() function is primarily used with Series objects to perform element-wise transformations or value mappings, such as replacing values based on a dictionary or applying a simple function to each element. It is limited to Series and works best for straightforward, element-by-element operations. On the other hand, apply() is more versatile and can be used on both Series and DataFrames. When applied to a DataFrame, it allows you to run functions across either rows or columns (by specifying the axis), enabling more complex, custom operations and aggregations. Thus, while map() is ideal for simple mappings on single columns, apply() is preferred for more advanced and flexible processing, especially when working with multiple columns or needing row-wise or column-wise computations.

21.Why is NumPy’s array slicing faster than Python’s list slicing?
- NumPy’s array slicing is faster than Python’s list slicing because NumPy arrays are stored as contiguous blocks of homogeneous memory, allowing slicing to create views of the original data without copying it. This means when you slice a NumPy array, it simply creates a new reference to the existing data segment, which is very efficient in both time and memory. In contrast, Python lists are collections of pointers to objects scattered in memory, so slicing a list creates a new list with copies of references, involving more overhead and memory allocation. Additionally, NumPy’s operations leverage low-level, optimized C code, further speeding up slicing and subsequent computations compared to Python’s interpreted list operations.

```python
# 4.How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample DataFrame
data = {
    'A': [1, 2, 3, 4, 5],
    'B': [5, 4, 3, 2, 1],
    'C': [2, 3, 4, 5, 6]
}
df = pd.DataFrame(data)

# Calculate correlation matrix
corr = df.corr()

# Plot heatmap of correlation matrix
sns.heatmap(corr, annot=True, cmap='coolwarm')

plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# 11.Perform matrix multiplication using NumPy.
import numpy as np

# Define two matrices
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

# Perform matrix multiplication
result = np.dot(A, B)

# Alternatively, you can use the @ operator in Python 3.5+
# result = A @ B

print("Matrix A:")
print(A)

print("\nMatrix B:")
print(B)

print("\nResult of A x B:")
print(result)
```

```python
# 13.Create a 3D scatter plot using Plotly.

import plotly.graph_objects as go

# Sample 3D data
x = [1, 2, 3, 4, 5]
y = [5, 6, 7, 8, 9]
z = [10, 11, 12, 13, 14]

# Create 3D scatter plot
fig = go.Figure(data=[go.Scatter3d(
    x=x,
    y=y,
    z=z,
    mode='markers',
    marker=dict(
        size=6,
        color=z,                # Set color based on z values
        colorscale='Viridis',  # Color scale
        opacity=0.8
    )
)])

# Customize layout
fig.update_layout(
    title='3D Scatter Plot',
    scene=dict(
        xaxis_title='X Axis',
        yaxis_title='Y Axis',
        zaxis_title='Z Axis'
    )
)

# Show plot
fig.show()
```
